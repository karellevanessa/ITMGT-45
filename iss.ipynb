```python
import requests
import time 
import matplotlib.pyplot as plt

latitude = []
longitude = []
timestamp = []

for i in range(100):
    link = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data = link.json()
    lat = data['latitude']
    latitude.append(lat)
    long = data['longitude']
    longitude.append(long)
    stamp = data['timestamp']
    timestamp.append(stamp)

    time.sleep(10)

plt.figure(figsize=(12,6)) 
plt.title("ISS Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.scatter(longitude, latitude)
plt.grid(True)
plt.show()
```

```python
import requests
import time 
import matplotlib.pyplot as plt

latitude = []
longitude = []
timestamp = []

for i in range(100):
    link = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data = link.json()
    lat = data['latitude']
    latitude.append(lat)
    long = data['longitude']
    longitude.append(long)
    stamp = data['timestamp']
    timestamp.append(stamp)

    time.sleep(10)

plt.figure(figsize=(12,6)) 
plt.title("ISS Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.plot(longitude, latitude, color='gray', linestyle = '-', linewidth = 2, marker = 'o', markersize = 5)
plt.grid(True)

for long, lat, point_label in zip(longitude,latitude,timestamp): #add labels to each scatter point
    plt.annotate(str(point_label), (long,lat), fontsize = 6, color = 'black')

plt.show()
```
